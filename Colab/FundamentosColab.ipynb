# Tutorial sobre cómo usar Google Colab en Física de la Radioterapia
-----
## Fundamentos

### ¿Qué es Google Colab?

Google Colab es un entorno de desarrollo y colaboración en la nube disponible gratuitamente para cualquier usuario de  una cuenta de Google.

Con nuestra cuenta de la UCM tenemos acceso a este entorno.

Se basa en cuadernos Jupyter, una implemtación en Python de un paradigma denominado **Computación Interactiva**. 

La Computación Interactiva permite juntar en el mismo documento código, resultados y texto.

- El código es interpretado, se puede modificar en el propio cuaderno sin tener que ser recompilado
- Los resultados se actualizan reflejando las modificaciones del código
- El texto permite explicar los resultados, hacer otras aclaraciones o documentar el código. Se escribe en formato markdown.

El formato markdown permite de forma flexible dar formato, estructurar, incluir fórmulas, tablas...

### ¿Cómo crear un cuaderno de Colab?

En Google Drive hay crear un nuevo documento que sea de tipo Google Colaboratory. 

Desde nuestro navegador en la página de Google Drive hacer `Nuevo > Más > Google Colaboratory`

Por defecto se crea un documento `Untitle0.ipynb` que podemos renombrar como queramos.

Para localizar el documento basta con ir a Recientes o buscar en gDrive. 

Aunque en gDrive hay una carpteta de Colab Notebooks los cuadernos podemos colocarnos donde queramos.

Alternativamente el cuaderno puede estar guardado en GitHub y abrirse desde un enlace que se proporcione.

### ¿Qué significa la extensión ipynb?

ipython notebook, la tecnología en la que se basan los cuadernos jupyter

### ¿Cómo colaborar en un cuaderno?

Como cualquier otro documento en gDrive el propietario puede administrar los permisos para ver, editar.

Varios usuarios pueden trabajar y modificar simultáneamente un documento si tienen los permisos adecuados.

### ¿Cómo trabajar sobre un cuaderno de otro sin alterar el original?

Un usuario puede hacer una copia de un cuaderno en su gDrive y comenzar a modificar esta copia.

Esta es la forma natural de trabajar cuando se accede al cuaderno cargándolo desde GitHub.

### ¿Cómo funciona?

Google Colab nos proporciona acceso a una máquina virtual con la que interaccionamos a través del cuaderno.

La máquina virtual corre sobre Ubuntu Linux pero en general esto será algo transparente para nosotros.

La máquina virtual se destruye cuando acabamos la sesión. Todo el software que hayamos tenido que instalar se perderá y tendrá que ser instalado de nuevo en una sesión posterior.

Que se destruya la máquina es independiente de que mantengamos la copia local del cuaderno en nuestro gDrive.

### ¿Cómo accedo a datos propios para analizarlos?

Los cuadernos Colab pueden acceder a ficheros almacenados en nuestro gDrive.

Para hacer esto primero es necesario montar gDrive mediante un código específico. Una vez montado gDrive el código que escribamos en Colab puede navegar por la estuctura de archivos de gDrive y cargar los datos que se necesiten.

### ¿Cómo se aumenta la funcionalidad de Colab?

Python ofrece un entorno fácilmente ampliable mediante módulos de software.

En un entorno genérico Python lo habitual es primero instalar el módulo de interés y después importarlo en el código.

El gestor de módulos más común en Python es `pip`

El entorno que por defecto ofrece Colab tiene muchos módulos ya instalados que son directamete importables en el código.

Si encontramos un `módulo` que necesitamos en nuestro cuaderno que no esté diponible por defecto en Colab siempre podemos instalarlo desde dentro de nuestro cuaderno mediante el comando mágico

`%pip install módulo`

Estas instalaciones se pierden al destruirse la máquina virtual. Se tienen que realizar cuando carguemos de nuevo el cuaderno y se nos asigne una nueva máquina.
